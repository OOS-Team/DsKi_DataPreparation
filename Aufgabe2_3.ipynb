```python
#URL
url = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States"
#Read Table
tables = pd.read_html(url)

# Wenn die Webseite mehr als eine Tabelle enthält, können Sie die gewünschte Tabelle mit ihrem Index auswählen:
# z.B., um die erste Tabelle zu wählen:
df = tables[1]
df.head()
```

```python
# Löschen der Obersten Spalten Namen
df.columns = df.columns.get_level_values(1)

df.head()
```

```python
df.info()
```

```python
#Benötigte Spalten Projezieren
state = df[["Flag, name and postal abbreviation[10]"]]
abbrev = df[["Flag, name and postal abbreviation[10].1"]]
capital = df[["Capital"]]
largest_pop_city = df[["Largest[14]"]]
established = df[["Ratification or admission[A]"]]
population = df[["Population (January 2023 estimate))[12]"]]
reps = df[["Number of Reps."]]

#Speichern aller qkm Werte in area
areas = df[["km2"]]

#Spaltennamen anpassen
areas.columns = ["total_area","land_area","water_area"]

#Restliche Spalten projezieren
total_area = areas[["total_area"]]
land_area = areas[["land_area"]]
water_area = areas[["water_area"]]
```

```python
# Data Frame keine Series
type(largest_pop_city)
```

```python
# Konvertiere die einzelnen DataFrame-Spalten in Series
state_series = state.iloc[:, 0]
abbrev_series = abbrev.iloc[:, 0]
capital_series = capital.iloc[:, 0]
largest_pop_city_series = largest_pop_city.iloc[:, 0]
established_series = established.iloc[:, 0]
population_series = population.iloc[:, 0]
total_area_series = total_area.iloc[:, 0]
land_area_series = land_area.iloc[:, 0]
water_area_series = water_area.iloc[:, 0]
reps_series = reps.iloc[:, 0]
```

```python
# Erstelle nun deinen neuen DataFrame
states = pd.DataFrame({
    "state": state_series,
    "abbrev": abbrev_series,
    "capital": capital_series,
    "largest_pop_city": largest_pop_city_series,
    "established": established_series,
    "population": population_series,
    "total_area": total_area_series,
    "land_area": land_area_series,
    "water_area": water_area_series,
    "# reps": reps_series
    
}, index=pd.RangeIndex(start=0, stop=len(state_series), step=1))

states.head()
```

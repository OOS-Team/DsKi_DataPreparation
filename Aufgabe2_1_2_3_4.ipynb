### 2.3
Was passiert, wenn sie map() auf Null-Werte anwenden?

Standardverhalten: Standardmäßig werden NaN-Werte von map() ignoriert und bleiben als NaN im resultierenden DataFrame oder der Serie erhalten. Das bedeutet, dass, wenn ein NaN-Wert auftritt, er nach der Anwendung von map() weiterhin als NaN vorhanden sein wird.

```python
import pandas as pd
import numpy as np

# Erstelle eine Serie mit einigen NaN-Werten
s = pd.Series([1, 2, np.nan, 4])

# Verwende map() mit einem Wörterbuch
mapped_s = s.map({1: 'one', 2: 'two', 4: 'four'})

print(mapped_s)
```

Beispiel mit Nan-Default wert

```python
import numpy as np

# Erstelle eine Serie mit einigen NaN-Werten
s = pd.Series([1, 2, np.nan, 4])

# Erstelle ein Mapping-Wörterbuch, das auch NaN-Werte berücksichtigt
mapping = {1: 'one', 2: 'two', 4: 'four', np.nan: 'missing'}

# Verwende map() mit dem angepassten Wörterbuch
mapped_s = s.map(mapping)

print(mapped_s)
```

### 3.2

Ermitteln Sie die Nachnamen der Passagiere (jeweils nur der erste Name).

```python

# Funktion, um den Vornamen zu extrahieren
def extrahiere_vornamen(name):
    match = re.search(r'^([^,]+),', name)
    if match:
        return match.group(1)  # Der Text nach dem Komma
    return None  # Rückgabe von None, falls kein Match vorhanden

# Erstelle eine neue Spalte 'Vornamen' durch Anwenden der Funktion
t['Nachname'] = t['Name'].apply(extrahiere_vornamen)

t.head(5)
```

- `^`: Dies markiert den Anfang des Strings.
- `([^,]+)`: 
  - Die eckigen Klammern `[ ]` definieren eine Zeichenklasse.
  - `^` innerhalb der eckigen Klammern bedeutet "nicht". Also `[^,]` entspricht "jedem Zeichen außer einem Komma".
  - Das Pluszeichen `+` bedeutet "eins oder mehr von den vorangehenden Elementen". In diesem Fall also "eins oder mehr von jedem Zeichen, das kein Komma ist".
- `,`: Das Komma am Ende des Musters stellt sicher, dass das Muster aufhört, sobald ein Komma im Text gefunden wird.

```python
# a) Extrahieren der Anrede
t['Title'] = t['Name'].str.extract(r'\b(Mr\.|Mrs\.|Miss|Ms\.)', expand=False)

t.head(25)
```

- `\b`: Dies ist ein Wortgrenzen-Anker, der sicherstellt, dass das Muster nur übereinstimmt, wenn es am Anfang eines Wortes steht. Dies verhindert, dass das Muster inmitten von Wörtern übereinstimmt.

- `(Mr\.|Mrs\.|Miss|Ms\.)`: Dies ist eine Gruppe, die aus mehreren Alternativen besteht, getrennt durch das Pipe-Zeichen `|`, was "oder" bedeutet. Das Muster übereinstimmt, wenn einer der folgenden Fälle im Text gefunden wird:
  - `Mr\.`: Die Anrede "Mr." (mit einem Punkt, da `\.` einen literalen Punkt darstellt, nicht irgendein Zeichen, wie es bei `.` der Fall wäre).
  - `Mrs\.`: Die Anrede "Mrs.".
  - `Miss`: Das Wort "Miss".
  - `Ms\.`: Die Anrede "Ms.".

b) Daraus kann man dann einen regulären Ausdruck definieren.

```python
# Regulärer Ausdruck
pattern = r'^(?P<Last>[^,]+), (?P<Title>Mr\.|Mrs\.|Miss|Ms\.) (?P<First>[^(]+)(?: \((?P<Maiden>[^)]+)\))?$'
```

1. `^(?P<Last>[^,]+),`:
   - `^`: Signalisiert den Anfang des Strings.
   - `(?P<Last>[^,]+)`: Benannte Gruppe `Last`, die alle Zeichen bis zum ersten Komma (`[^,]+`) erfasst. `[^,]` bedeutet jedes Zeichen außer einem Komma, und `+` bedeutet "eins oder mehr davon".
<br>
<br>

2. `(?P<Title>Mr\.|Mrs\.|Miss|Ms\.)`:
   - `(?P<Title>...)`: Benannte Gruppe `Title`, die eine der Anreden erfasst. Es kann `Mr.`, `Mrs.`, `Miss` oder `Ms.` sein. Das `\.` steht für einen tatsächlichen Punkt, da ein Punkt in regulären Ausdrücken jedes Zeichen darstellen kann.
<br>
<br>

3. `(?P<First>[^(]+)`: 
   - `(?P<First>...)`: Benannte Gruppe `First`, die alle Zeichen nach dem Titel und vor einer öffnenden Klammer `(` (falls vorhanden) erfasst. `[^(]+` bedeutet jedes Zeichen außer einer öffnenden Klammer, und `+` bedeutet wieder "eins oder mehr davon".
<br>
<br>

4. `(?: \((?P<Maiden>[^)]+)\))?`:
   - `(?:...)`: Ein nicht erfassender Gruppierungsteil, der für die Klammerung im Ausdruck verwendet wird, aber nicht als separate Gruppe erfasst wird.
   - `\((?P<Maiden>[^)]+)\)`: Eine optionale benannte Gruppe `Maiden`, die innerhalb von Klammern stehende Zeichen erfasst. `[^)]+` bedeutet jedes Zeichen außer einer schließenden Klammer, und `+` bedeutet "eins oder mehr davon".
   - `?` am Ende macht den gesamten Klammerausdruck optional.

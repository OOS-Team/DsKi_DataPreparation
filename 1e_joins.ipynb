# Joins

```python
import pandas as pd
%xmode minimal
pd.set_option("mode.copy_on_write", True)
```

```python
file_prep = 'data/titanic_prepared.pickle'
t = pd.read_pickle(file_prep)
t.sample(3)
```

Wir machen Texte für die Klassen um eine Beschreibung hinzuzufügen

```python
t["Class"].unique()
```

```python
data={
    "key" : ['3','2', '1', 'V', 'R', 'E', 'D', 'B'],
    "Klasse": ['3rd class', '2nd', 'first', 'cooks', 'restaurant', 'Technik', 'Sailor', 'Musik'],
    "Status": ["Passagier", "Passagier", "Passagier", "Crew", "Crew", "Crew", "Crew", "Crew"]
}

klassen =pd.DataFrame(data).convert_dtypes()
klassen
```

```python
t = t[["Name", "Sex", "Class", "Age", "Joined", "Price"]]
```

Manipulation der Daten, sodass wir auch NULL Werte in der Class haben.
Und wir wollen auch einen ungültigen Wert haben der keinen Gültigen Partner in der Klasse findet

```python
t.loc[[2021, 888, 1890],"Class"] = pd.NA
t.loc[[2021, 888, 1890]]
t.loc[[2021, 888, 1980]]
```

## Inner JOIN

Ist equal Join (Bedingung zB SpalteA==SpalteB oder wie im folgenden Bsp Class==Key )

```python
t2 = t.merge(klassen, how='inner', left_on="Class", right_on="key")
t2
```

```python
print(f'Shape von t :  {t.shape}  und Shape von t2:  {t2.shape}')
```

### Outer Join
damit wir bei drei Passagieren ohne `Class` nicht verlieren.

```python
(t.shape, t2.shape)
```

```python
t2 = t.merge(klassen, how='left', left_on="Class", right_index=True, validate='m:1')
t2.loc[t2["Class"].isna()]
```

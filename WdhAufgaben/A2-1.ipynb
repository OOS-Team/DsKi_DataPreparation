```python
def hafen_setzten(c):

    if c == embarked_names[0][0]: return embarked_names[0]
    if c == embarked_names[1][0]: return embarked_names[1]
    if c == embarked_names[2][0]: return embarked_names[2]
    if c == embarked_names[3][0]: return embarked_names[3]
```

```python
t['Embarked'] = t['Embarked'].map(hafen_setzten)
```

## 2.7
Konzept: <br>
Chained Kommandos lassen sich durch `()` erstellen <br>
Funktion: <br>
Lamda mit mehreren if´s : `lambda c: 'F' if c == 'female' else('M' if c== 'male' else 'X')`

```python
t.loc[t['Pclass']==1, 'Sex'] = (t.loc[t['Pclass']==1, 'Sex']
    .apply(lambda c: 'F' if c == 'female' else('M' if c== 'male' else 'X')))
```

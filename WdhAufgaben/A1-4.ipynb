4.7 <br>
Wert in Spalte einfügen nach einer Bedingung: <br>
`wetter.loc[(wetter["Regen"]>0) & (wetter["Sonne"] > 0), "Text"] = "April"` <br>
Werte in Spalte einfügen nach Bedingung und überprüfung auf NA Werte: <br>
`wetter.loc[(wetter["Sonne"] >0) & (wetter["Text"].isna()), "Text"] = "Sonne"`

```python
wetter["Text"] = pd.NA

wetter.loc[(wetter["Regen"]>0) & (wetter["Sonne"] > 0), "Text"] = "April"
wetter.loc[(wetter["Sonne"] >0) & (wetter["Text"].isna()), "Text"] = "Sonne"
wetter.loc[(wetter["Regen"]> 0) & (wetter["Text"].isna()), "Text"] = "Regen"
```

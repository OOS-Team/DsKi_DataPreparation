```python
t['Titel']= t['Name'].str.extract(r'([,]\s*\w+.?)')  
```

```python
t.head()
```

```python
def titelGender(row):
    sex = row['Sex']
    parch = row['Parch']
    
    if sex == 'male': return 'Mr.'
    if (sex == 'female') & (parch==0): return 'Miss. '
    if (sex == 'female') & (parch==1): return 'Mrs. '
```

```python
t['Gender-Titel'] = t.apply(titelGender, axis=1)
```

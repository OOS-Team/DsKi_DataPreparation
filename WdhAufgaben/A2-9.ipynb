## Aufgabe 9.1
<br> Funktionen: <br>


Nach Anfangsbuchstaben extrahieren: `col.startswith("A")` <br>
Spalten löschen: `b.drop("`Spalte`")` <br>
Datum ISO Format: `pd.to_datetime(`b["Spalte"]`)` <br>
Mit Zeiträumen rechnen: `pd.to_timedelta(`b['Woche']`, unit='w')` <br> UNITS:  

M - Monate, 
Y - Jahre, 
w - Wochen, 
d - Tage, 
h - Stunden, 
m oder min - Minuten, 
s - Sekunden

```python
b = pd.read_csv('../data/Aufgaben/billboard.csv')
```

```python
b.sample(5)
```

```python
wk_columns = [col for col in b.columns if col.startswith('wk')]
```

```python
b['Woche'] = b[wk_columns].notna().sum(axis=1)
```

```python
b = b.drop(columns=wk_columns)
```

```python
b =b.rename(columns={'date.entered': 'start-date'})
```

```python
b.head()
```

```python
b['end-date'] = pd.to_datetime(b['start-date']) + pd.to_timedelta(b['Woche'], unit='w')
```

```python
b.sample(5)
```

```python
ts = pd.read_csv('../data/Aufgaben/country_timeseries.csv')
```

```python
ts.sample(3)
```

```python
ts_cases = ts.melt(id_vars=['Date', 'Day'], var_name='Countrys_Cases', value_name='Number of Cases', value_vars=[col for col in ts.columns if 'Cases' in col] )
```

```python
ts_deaths = ts.melt(id_vars=['Date', 'Day'], var_name='Countrys_Death', value_name='Number of Deaths', value_vars=[col for col in ts.columns if 'Deaths_' in col] )
```

```python
ts_deaths
```

```python
ts = ts_cases.merge(ts_deaths, on=['Date', 'Day'])
```

```python
ts.head()
```

```python
ts['Deaths-per-Case'] = ts['Number of Cases'] / ts['Number of Deaths']
```

```python
ts['Deaths-per-Case'].notna().sum()
```

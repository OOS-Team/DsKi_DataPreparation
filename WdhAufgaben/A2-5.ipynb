```python
import pandas as pd
import os
%xmode minimal
```

```python
url = 'https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv'
file = '../data/titanic_2_original.pickle' # Damit es local gespeichert wird

if os.path.exists(file):
    t = pd.read_pickle(file)
else:
    t = pd.read_csv(url)
    t.to_pickle(file)
t.sample(10)
```

```python
import seaborn as sns
```

```python
sns.histplot(data=t, x=t['Age'])# , bins=10)  # KDE (Kernel Density Estimate) auf False gesetzt, bins f√ºr die Anzahl der Balken
```

```python
sns.histplot(data=t, x=t['Age'], hue=t['Sex'])
```

```python
sns.displot(data=t, x='Age', col='Sex')
```

```python
sns.histplot(data=t, x=t['Age'], hue=t['Survived'])
```

```python
sns.displot(data=t, x='Age', col='Survived')
```

```python
sns.histplot(data=t, x=t['Age'], hue=t['Pclass'])
```

```python
sns.displot(data=t, x='Age', col='Pclass')
```

### 7.1. H2020 Projekte

```python
import pandas as pd
%xmode minimal
pd.set_option("mode.copy_on_write", True)
```

```python
projects_path = "./data/cordis_eu_2020/project.csv"
org_path = "./data/cordis_eu_2020/organization.csv"
```

a) Laden Sie die entsprechenden CSV Dateien
Beachten Sie, dass die Behörde europäisch ist und setzen Sie daher beim Laden der
CSV Datei das CSV-Trennzeichen und Dezimaltrennzeichen entsprechend.

```python
eu_proj = pd.read_csv(projects_path, sep=';', decimal=',', dtype_backend='numpy_nullable')
eu_org = pd.read_csv(org_path, sep=';', decimal=',', dtype_backend='numpy_nullable')
```

```python
eu_org.head(3)
```

```python
eu_proj.head(3)
```

b) Wie viele Projekte und Organisationen im Programm H2020 gibt es?
In den beiden Dateien steht jeweils die „id“ für den Schlüssel der Organisation (org id)
bzw. des Projektes

```python
print(f'Es gibt {eu_org.shape[0]} Organisationen.')
print(f'Es gibt {eu_proj.shape[0]} Projekte.')
```

c) Joinen Sie die beiden Dateien über die Projekt ID.

```python
eu_proj = eu_proj.rename(columns={"id":"projectID"})
```

c) Joinen Sie die beiden Dateien über die Projekt ID.

```python
eu_proj.head(3)
```

```python
eu_org.head(3)
```

```python
eu_complete = eu_org.merge(eu_proj, how='inner', left_on='projectID', right_on='projectID')
eu_complete.head(3)
```

```python
eu_complete.columns
```

d) Welche Organisation hat das größte Budget (Summe aller Projekte) und wie hoch ist es?

```python
# Gruppieren der Daten nach 'name' oder 'organisationID' und Berechnung der Summe der 'totalCost_y' für jede Gruppe
grouped_data = eu_complete.groupby('name')['totalCost_y'].sum()

# Finden der Organisation mit dem höchsten Gesamtbudget
organisation_with_max_budget = grouped_data.idxmax()
max_budget = grouped_data.max()

print(f"Die Organisation mit dem größten Budget ist {organisation_with_max_budget} mit einem Budget von {max_budget} Euro.")
```

e) Wie viele Zeilen wurden nicht erfolgreich gejoint?

```python
# Anzahl der Zeilen in den ursprünglichen DataFrames
num_rows_eu_org = len(eu_org)
num_rows_eu_proj = len(eu_proj)

# Anzahl der Zeilen nach dem inneren Join
num_rows_eu_complete = len(eu_complete)

print(f'Anzahl der Zeilen in Organisation {num_rows_eu_org}. Anzahl der Zeilen in Projekt {num_rows_eu_proj}. Anzahl nach Join {num_rows_eu_complete}.')
```

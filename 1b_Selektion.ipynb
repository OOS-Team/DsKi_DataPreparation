### Slicing in Python

```python
liste = [3,34,543,43,234,1,65,78]
print(liste[2])
print(liste[2:3])
print(liste[2:])
print(liste[:3])
print(liste[:])
print(liste[:-1])
```

### Slicing in Pandas

```python
df2.loc[100001:100003]
```

### Neuen Index anlegen und alten löschen

```python
df2 = df2.reset_index().set_index("OrderItem")
df2.head(3)
```

```python
df2.loc[20]
```

```python
# Geht nicht da nicht unique und unsortiert
df2.loc[20:30]
```

```python
# Geht da Sortiert
df2 = df2.sort_index()
df2.loc[20:30]
```

Teste, ob Index sortiert ist

```python
assert(df2.index.is_monotonic_increasing)
```

## Multiindex

```python
gbi = df2.reset_index().set_index(["Country", "City", "Date"])
gbi.head()
```

```python
# Index ist nicht sortiert
gbi.index
```

```python
gbi.loc['DE', 'Hannover', '2011-06-07']
```

```python
print(gbi.index.is_unique)
```

```python
print(gbi.index.is_monotonic_decreasing)
```

Slicing bei Multiindex

```python
gbi.loc[:, 'Hannover', '2011-06-01':'2011-06-07']
```

```python
gbi = gbi.sort_index()
```

```python
gbi.loc[:, 'Hannover', '2011-06-01':'2011-06-07']
```

Jetzt wollen wir für die selektierten Zeilen ein Update machen <br>
und zwar: Division = 'Socken'

```python
gbi.loc[:, ['Hannover', 'Berlin'], '2011-06-07']
```

```python
# Geht leider nicht
gbi.loc[:, ['Hannover', 'Berlin'], '2011-06-07']["Division"] = 'Socken'
```

```python
gbi.loc[('DE', ['Hannover', 'Berlin'], '2011-06-07'),"Division"] = 'Socken'
```

```python
# Testen des Ergibnis - braucht eigentlich noch weiteres Vorwissen
gbi.loc[gbi["Division"] == 'Socken']
```

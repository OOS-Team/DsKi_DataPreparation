# Standard Einstellung

```python
import pandas as pd
import os
%xmode minimal
pd.set_option("mode.copy_on_write", False) # Um seaborn richtig zu benutzen False
```

# Daten laden

```python
url = 'https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv'
file = 'data/titanic_2_original.pickle' # Damit es local gespeichert wird

if os.path.exists(file):
    t = pd.read_pickle(file)
else:
    t = pd.read_csv(url)
    t.to_pickle(file)
t.sample(3)
```

# Plotting

```python
import seaborn as sns
```

```python
t.dtypes
```

```python
t.astype({"Age":"float64"})
```

```python
g = sns.displot(data=t, x="Age")
```

```python
g = sns.displot(data=t, x="Age",col="Pclass",binwidth=10, kde=True) # binwitdth ist die breite der bins
g.set_axis_labels('Age bins', '#Passengers')
# g.savefig('Altersverteilung.png', format='png', dpi=150) #dpi=dotsPerInch
```

```python
bins = [0, 12, 18, 67, 100]
labels = ['baby', 'teenager', ' worker', 'elder']
t["Age class"] = pd.cut(t["Age"], bins=bins)
```

```python
sns.catplot(kind="bar", data=t, x="Age class", y="Survived", errorbar=('ci',False), estimator='sum', color='lightblue')
```

```python
t.groupby(by="Age class")["Survived"].sum()
```

```python
sns.lineplot(data=t.loc[t['Age']>=60], x='Age', y='Fare', estimator='mean', errorbar=None)
```

```python
sns.catplot(kind="box", data=t, y="Age", x="Pclass")
```

### 2.3
Was passiert, wenn sie map() auf Null-Werte anwenden?

Standardverhalten: Standardmäßig werden NaN-Werte von map() ignoriert und bleiben als NaN im resultierenden DataFrame oder der Serie erhalten. Das bedeutet, dass, wenn ein NaN-Wert auftritt, er nach der Anwendung von map() weiterhin als NaN vorhanden sein wird.

```python
import pandas as pd
import numpy as np

# Erstelle eine Serie mit einigen NaN-Werten
s = pd.Series([1, 2, np.nan, 4])

# Verwende map() mit einem Wörterbuch
mapped_s = s.map({1: 'one', 2: 'two', 4: 'four'})

print(mapped_s)
```

Beispiel mit Nan-Default wert

```python
import numpy as np

# Erstelle eine Serie mit einigen NaN-Werten
s = pd.Series([1, 2, np.nan, 4])

# Erstelle ein Mapping-Wörterbuch, das auch NaN-Werte berücksichtigt
mapping = {1: 'one', 2: 'two', 4: 'four', np.nan: 'missing'}

# Verwende map() mit dem angepassten Wörterbuch
mapped_s = s.map(mapping)

print(mapped_s)
```

### 2.6
Sie wollen basierend auf „Age“ jeweils einen Satz ausgeben. Beispiel: „Ich bin 33
Jahre alt.“ ausgeben. Vermeiden Sie den Satz „Ich bin nan Jahre alt.

```python
# Funktion, um einen Satz basierend auf dem Alter zu erstellen
def age_sentence(age):
    if pd.isna(age):  # Überprüfen, ob der Wert NaN ist
        return None  # Keine Ausgabe für NaN-Werte
    else:
        return f"Ich bin {int(age)} Jahre alt."

# Anwenden der Funktion auf jede Zeile der Spalte 'Age'
sentences = t['Age'].apply(age_sentence)

# Ausgeben der Sätze
for sentence in sentences:
    if sentence is not None:
        print(sentence)
```

### 3.2

Ermitteln Sie die Nachnamen der Passagiere (jeweils nur der erste Name).

```python

# Funktion, um den Vornamen zu extrahieren
def extrahiere_vornamen(name):
    match = re.search(r',\s*(.*)', name)
    if match:
        return match.group(1)  # Der Text nach dem Komma
    return None  # Rückgabe von None, falls kein Match vorhanden

# Erstelle eine neue Spalte 'Vornamen' durch Anwenden der Funktion
t['Vornamen'] = t['Name'].apply(extrahiere_vornamen)

t.head(5)
```
